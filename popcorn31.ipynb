```python
#POPCORN
# Function to create a user profile
def create_user_profile():
    # Store user information in variables
    first_name = input("Enter your first name: ")
    last_name = input("Enter your last name: ")
    age = input("Enter your age: ")
    email = input("Enter your email address: ")
    hobby = input("Enter your favorite hobby: ")

    # Concatenate the information to create a profile
    profile = (
        f"Profile Information:\n"
        f"Name: {first_name} {last_name}\n"
        f"Age: {age}\n"
        f"Email: {email}\n"
        f"Hobby: {hobby}"
    )

    # Print the user profile
    print("\n" + profile)

# Call the function to create a user profile
create_user_profile()
```

```python
#HOMEWORK
# Profile Creation Script
def create_profile(name, age, location, interests=None, skills=None, experience=None):
    profile = {
        "Name": name,
        "Age": age,
        "Location": location,
        "Interests": interests if interests else [],
        "Skills": skills if skills else {},
        "Experience": experience if experience else []
    }
    
    return profile

# Example profiles
profile_jane = create_profile(
    name="Jane Doe",
    age=28,
    location="New York",
    interests=["Hiking", "Reading", "Photography"],
    skills={"Programming": ["Python", "JavaScript", "HTML/CSS"], "Languages": ["English", "Spanish"]},
    experience=[
        {"role": "Software Developer", "company": "Tech Solutions", "years": 3},
        {"role": "Web Designer", "company": "Creative Studio", "years": 2}
    ]
)

profile_john = create_profile(
    name="John Smith",
    age=35,
    location="San Francisco",
    interests=["Cycling", "Cooking"],
    skills={"Data Analysis": ["SQL", "R", "Python"], "Project Management": ["Agile", "Scrum"]},
    experience=[
        {"role": "Data Analyst", "company": "Data Corp", "years": 4},
        {"role": "Project Manager", "company": "Innovate Inc", "years": 3}
    ]
)

# Display profiles
def display_profile(profile):
    print(f"Name: {profile['Name']}")
    print(f"Age: {profile['Age']}")
    print(f"Location: {profile['Location']}")
    print("Interests: ", ", ".join(profile["Interests"]) if profile["Interests"] else "None")
    print("\nSkills:")
    for skill, details in profile["Skills"].items():
        print(f" - {skill}: {', '.join(details)}")
    print("\nExperience:")
    if profile["Experience"]:
        for job in profile["Experience"]:
            print(f" - {job['role']} at {job['company']} for {job['years']} years")
    else:
        print("No experience listed.")
    print("\n")

# Display example profiles
display_profile(profile_jane)
display_profile(profile_john)
```

```python
import random
import time

# Function to generate a unique ID based on timestamp and random number
def generate_unique_id():
    timestamp = int(time.time() * 1000)  # Current time in milliseconds
    random_num = random.randint(0, 9999)  # Random number from 0 to 9999
    return f"ID{timestamp}-{random_num}"

# Function to create a profile with various attributes
def create_profile(name, age, location, hobbies=None, languages=None, experience=None):
    if hobbies is None:
        hobbies = []
    if languages is None:
        languages = []
    if experience is None:
        experience = []
    
    profile_id = generate_unique_id()
    
    profile = {
        "id": profile_id,
        "name": name,
        "age": age,
        "location": location,
        "hobbies": hobbies,
        "languages": languages,
        "experience": experience
    }
    
    return profile

# Store profiles in a list
profiles = []

# Create sample profiles
profile_alice = create_profile(
    "Alice Johnson",
    30,
    "Los Angeles",
    ["Painting", "Yoga", "Reading"],
    ["English", "French"],
    [
        {"role": "Graphic Designer", "company": "Creative Solutions", "years": 4},
        {"role": "Art Director", "company": "Design Studio", "years": 2}
    ]
)

profile_bob = create_profile(
    "Bob Williams",
    42,
    "Chicago",
    ["Golf", "Cooking"],
    ["English", "Italian"],
    [
        {"role": "Chef", "company": "Gourmet Kitchen", "years": 5},
        {"role": "Restaurant Manager", "company": "Fine Dine", "years": 3}
    ]
)

# Add profiles to the profiles list
profiles.extend([profile_alice, profile_bob])

# Function to display profile information
def display_profiles(profiles_list):
    for profile in profiles_list:
        print(f"Profile ID: {profile['id']}")
        print(f"Name: {profile['name']}")
        print(f"Age: {profile['age']}")
        print(f"Location: {profile['location']}")
        
        print("Hobbies:")
        for hobby in profile['hobbies']:
            print(f" - {hobby}")

        print("Languages:")
        for language in profile['languages']:
            print(f" - {language}")

        print("Experience:")
        for job in profile['experience']:
            print(f" - {job['role']} at {job['company']} for {job['years']} years")
        
        print("\n")

# Display all profiles
display_profiles(profiles)
```
